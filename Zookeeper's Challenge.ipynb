```python
#Import Modules
from datetime import date
```

```python
# animal.py

# creating a template for the animals
class Animal:
    numOfAnimals = 0

    #using __init__ to automatically call. Setting the values to self for each animal
    def __init__(self, age, species, name, animal_id, birth_date, color, sex, weight, originating_zoo, date_arrival):
        self.species = species
        self.name = name
        self.animal_id = animal_id
        self.birth_date = birth_date
        self.color = color
        self.sex = sex
        self.weight = weight
        self.originating_zoo = originating_zoo
        self.date_arrival = date_arrival
        self.age = age

        #keeping track of how many have been created/added
        Animal.numOfAnimals += 1 
        
```

```python
# hyena.py

# using the animal template to call the specific animals info
class Hyena(Animal):
    numOfHyenas = 0
    list_of_hyena_names = [] #keeping track of the names

    #reading in the names and parsing out the the specific animals with Lines[number].split
    file_path = 'animalNames.txt'
    with open(file_path, "r") as file:
        lines = file.readlines()
        list_of_hyena_names.extend(lines[2].strip().split(', ')) # line[2] = hyena names

    # assigning the values
    def __init__(self, age="a_age", name="a_name", animal_id="an_id", birth_date="2099-01-01", color="a_color", sex="a_sex", weight="a_weight", originating_zoo="a_zoo", date_arrival="2099-01-01"):

        #calling the animal class with super to stay more accurate
        super().__init__(age, "hyena", name, animal_id, birth_date, color, sex, weight, originating_zoo, date_arrival)  
        
        #just keeping track of the amount of times I run this since it changed the names
        Hyena.numOfHyenas += 1 

    # making sure there is a list and taking out each name so it doesn't repeat
    def get_hyena_name(self):  # using self so I dont have to fill it in later
        if Hyena.list_of_hyena_names:
            return Hyena.list_of_hyena_names.pop(0)
        return "Unnamed Hyena" # just in case I run it too many times without resetting
```

```python
# lion.py
class Lion(Animal):
    numOfLions = 0
    lion_sound = "Roarrr"
    list_of_lion_names = []

    file_path = 'animalNames.txt'
    with open(file_path, "r") as file:
        lines = file.readlines()
        if lines:
            list_of_lion_names.extend(lines[6].strip().split(', ')) # line[6] = lion names

    def __init__(self, age="a_age", name="a_name", animal_id="an_id", birth_date="2099-01-01", color="a_color", sex="a_sex", weight="a_weight", originating_zoo="a_zoo", date_arrival="2099-01-01"):

        super().__init__(age, "lion", name, animal_id, birth_date, color, sex, weight, originating_zoo, date_arrival)  
        Lion.numOfLions += 1

    def get_lion_name(self):  
        if Lion.list_of_lion_names:
            return Lion.list_of_lion_names.pop(0)
        return "Unnamed Lion"
```

```python
# tiger.py
class Tiger(Animal):
    numOfTigers = 0
    tiger_sound = "Mew"
    list_of_tiger_names = []

    file_path = 'animalNames.txt'
    with open(file_path, "r") as file:
        lines = file.readlines()
        if len(lines) >= 2:
            list_of_tiger_names.extend(lines[14].strip().split(', '))  # line[14] = tiger names

    def __init__(self, age="a_age", name="a_name", animal_id="an_id", birth_date="2099-01-01", color="a_color", sex="a_sex", weight="a_weight", originating_zoo="a_zoo", date_arrival="2099-01-01"):

        super().__init__(age, "tiger", name, animal_id, birth_date, color, sex, weight, originating_zoo, date_arrival)  
        Tiger.numOfTigers += 1

    def get_tiger_name(self):  
        if Tiger.list_of_tiger_names:
            return Tiger.list_of_tiger_names.pop(0)
        return "Unnamed Tiger"
```

```python
# bear.py
class Bear(Animal):
    numOfBear = 0
    bear_sound = "grmmm"
    list_of_bear_names = []

    file_path = 'animalNames.txt'
    with open(file_path, "r") as file:
        lines = file.readlines()
        if len(lines) >= 2:
            list_of_bear_names.extend(lines[10].strip().split(', '))  # line[10] = bear names

    def __init__(self, age="a_age", name="a_name",animal_id="an_id", birth_date="2099-01-01", color="a_color", sex="a_sex", weight="a_weight", originating_zoo="a_zoo", date_arrival="2099-01-01"):

        super().__init__(age, "bear", name, animal_id, birth_date, color, sex, weight, originating_zoo, date_arrival)  
        Bear.numOfBear += 1

    def get_bear_name(self):  
        if Bear.list_of_bear_names:
            return Bear.list_of_bear_names.pop(0)
        return "Unnamed Bear"
```

```python
# main.py

# Lists for keeping track of values
list_of_hyenas = []
list_of_lions = []
list_of_tigers = []
list_of_bears = []

# creating sounds for later print (From the video, I thought it was cute)
hyena_sound = "Hehe"
lion_sound = "Roarrr"
tiger_sound = "Mew"
bear_sound = "grmmm"

# setting the dates from calculating birthday and current year for arrival
current_date = date.today()
current_year = current_date.year

# function that translate birthday season into an actual date
def calculate_birth(season, year):
    year_of_birth = int(current_year) - int(year) # simple math for age and current year

    season = season.lower()
    if "spring" in season:
        return str(year_of_birth) + "-03-01"
    elif "summer" in season:
        return str(year_of_birth) + "-06-01"
    elif "fall" in season:
        return str(year_of_birth) + "-09-01"
    elif "winter" in season:
        return str(year_of_birth) + "-12-01"
    else:
        return str(year_of_birth) + "-01-01"

# This is where we we split each of the sets we need, I put it in a def so i can tweak it easier than inside the if statement
def process_one_line(one_line):
    
    # splitting each groups into the words I am looking for
    # [age(as a number), sex, species]
    groups_of_words = one_line.strip().split(",")
    single_words = groups_of_words[0].strip().split(" ") 
    age_in_years = single_words[0] # the number for age
    a_sex = single_words[3] # the sex
    a_species = single_words[4] #the species
    
    # [season]
    single_words_g1 = groups_of_words[1].strip().split(" ")
    season = single_words_g1[2] # the season

    # [color]         
    single_words_g2 = groups_of_words[2].strip().split(", ")
    color = single_words_g2[0].strip() # the color

    # [weight]       
    single_words_g3 = groups_of_words[3].strip().split(" ")
    weight = single_words_g3[0].strip() # the number for weight 

    # [from zoo]
    from_zoo = groups_of_words[4].strip() + ", " + groups_of_words[5].strip() # combined from and region
    
    #using out birthdate calculator
    birth_date = calculate_birth(season, age_in_years)

    # creating the objects to be used later for the print
    # checking through each of the species to create the correct object
    if "hyena" in a_species.lower():
        # using the functions created earlier for each species
        my_hyena = Hyena(age_in_years, "", "", birth_date, color, a_sex, weight, from_zoo, current_date) 
        my_hyena.name = my_hyena.get_hyena_name() #calling the function
        my_hyena.animal_id = "Hy" + str(Hyena.numOfHyenas).zfill(2) # creating the unique id
        list_of_hyenas.append(my_hyena) # adding the object to the list
        
    if "bear" in a_species.lower():
        my_bear = Bear(age_in_years, "", "", birth_date, color, a_sex, weight, from_zoo, current_date)
        my_bear.name = my_bear.get_bear_name()
        my_bear.animal_id = "Be" + str(Bear.numOfBear).zfill(2)
        list_of_bears.append(my_bear)

    if "lion" in a_species.lower():
        my_lion = Lion(age_in_years,"", "", birth_date, color, a_sex, weight, from_zoo, current_date)
        my_lion.name = my_lion.get_lion_name()
        my_lion.animal_id = "Li" + str(Lion.numOfLions).zfill(2)
        list_of_lions.append(my_lion)
        
    if "tiger" in a_species.lower():
        my_tiger = Tiger(age_in_years, "", "", birth_date, color, a_sex, weight, from_zoo, current_date)
        my_tiger.name = my_tiger.get_tiger_name()
        my_tiger.animal_id = "Ti" + str(Tiger.numOfTigers).zfill(2)
        list_of_tigers.append(my_tiger)

# reading the file and running the function made earlier
file_path = "arrivingAnimals.txt"
with open(file_path, "r") as file:
    for line in file:
        process_one_line(line)

# Printing the number counts to check usage
print(f'Number of animals created: {Animal.numOfAnimals}')
print(f'Number of hyenas created: {Hyena.numOfHyenas}')
print(f'Number of lions created: {Lion.numOfLions}')
print(f'Number of tigers created: {Tiger.numOfTigers}')
print(f'Number of bears created: {Bear.numOfBear}')

# Printing the habitats using what was created in a clean easy to read way
print("\nHyenas Habitat:\n")
for hyena in list_of_hyenas:
    print(f"{hyena.animal_id}, {hyena.age} years old, {hyena.name}; birthdate: {hyena.birth_date}; {hyena.color}; " f"{hyena.sex}; {hyena.weight} pounds; Laugh: {hyena_sound}; {hyena.originating_zoo}; arrived: {hyena.date_arrival}")

print("\nLions Habitat:\n")
for lion in list_of_lions:
    print(f"{lion.animal_id}, [{lion.age}] years old, {lion.name}; birthdate: {lion.birth_date}; {lion.color}; " f"{lion.sex}; {lion.weight} pounds; Roar: {lion_sound}; {lion.originating_zoo}; arrived: {lion.date_arrival}")
    
print("\nTigers Habitat:\n")
for tiger in list_of_tigers:
    print(f"{tiger.animal_id}, {tiger.age} years old, {tiger.name}; birthdate: {tiger.birth_date}; {tiger.color}; " f"{tiger.sex}; {tiger.weight} pounds; Roar: {tiger_sound}; {tiger.originating_zoo}; arrived: {tiger.date_arrival}")

print("\nBears Habitat:\n")
for bear in list_of_bears:
    print(f"{bear.animal_id}, {bear.age} years old, {bear.name}; birthdate: {bear.birth_date}; {bear.color}; " f"{bear.sex}; {bear.weight} pounds; Roar: {bear_sound}; {bear.originating_zoo}; arrived: {bear.date_arrival}")
    
```
